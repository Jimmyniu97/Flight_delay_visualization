```python
import pandas as pd
import os
```

```python
def data_process(file_path):
    f=pd.read_csv(file_path)
    df=f[['ORIGIN','DEP_DELAY','ARR_DELAY']]#only select the origin and arrive delay time attribute 
    df=df.rename(columns={'ORIGIN':'code'})#rename the origin as code
    df['Total_delay']=df['DEP_DELAY']+df['ARR_DELAY']
    df_sub1=df[df['Total_delay']>0]#select the airport with delay
    df_sub1=df_sub1.groupby("code").size().reset_index(name='count')
    df_sub1['number']=df_sub1['count']
    df_sub1['count']=(df_sub1['count']-df_sub1['count'].mean())/(df_sub1['count'].std())
    df_sub1['count']=df_sub1['count'].abs()
    f1=pd.read_json("/Users/niuyuefeng/Desktop/Penn state course/4/Ds 330/Final project/Data/Raw_data/airports.json")
    f1=f1[f1['country']=='United States']
    f1=f1[f1['type']=='Airports']
    f1=f1[f1['state']!='Puerto Rico']
    f1=f1[f1['state']!='American Samoa']
    f1=f1[f1['state']!='Northern Mariana Islands']
    f1=f1[f1['state']!='US Virgin Islands']
    f1=f1[f1['state']!='Guam']
    df2=f1[['code','lat','lon','name','city']]
    data=df_sub1.set_index('code').join(df2.set_index('code'))
    data=data.dropna()#drop the Null value
    data.to_csv(r"/Users/niuyuefeng/Desktop/Penn state course/4/Ds 330/Final project/Data/bubble_data_2018.csv")
```
