```python
import pandas as pd
import numpy as np
```

```python
f=pd.read_csv("/Users/niuyuefeng/Desktop/Penn state course/4/Ds 330/Final project/Data/Raw_data/2009.csv")#read the delay airport files
```

```python
df=f[['ORIGIN','DEP_DELAY','ARR_DELAY']]#only select the origin and arrive delay time attribute 
df=df.rename(columns={'ORIGIN':'code'})#rename the origin as code
df['Total_delay']=df['DEP_DELAY']+df['ARR_DELAY']
```

```python
df
```

```python
df_sub1=df[df['Total_delay']>0]#select the airport with delay
df_sub1=df_sub1.groupby("code").size().reset_index(name='count')#count the delay times for the delayed airport
df_sub1
```

```python
df_sub1.describe()#since the data range is too large, we need to normalize it
```

```python
df_sub1['number']=df_sub1['count']
df_sub1
```

```python
df_sub1['count']=(df_sub1['count']-df_sub1['count'].mean())/(df_sub1['count'].std())
df_sub1['count']=df_sub1['count'].abs()
df_sub1#normalize the count of the delays and convert to positive numbers
```
